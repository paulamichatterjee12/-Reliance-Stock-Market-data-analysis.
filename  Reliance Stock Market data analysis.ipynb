## 3. Exploratory Data Analysis(EDA)

### Closing Price:- 
#### The closing price in stock market data analysis refers to the last traded price of a particular stock or at the end of a trading day. It is the final price at which a stock is traded before the market closes for the day.Analyzing closing prices is a common practice in technical analysis and financial research.

### 3.1.showing closing price over time using line plot.

```python
# Plotting the closing price over time
plt.figure(figsize=(12, 6))
plt.plot(stock_data['Date'], stock_data['Close'], label='Closing Price', color='blue')
plt.title('Reliance Industries Stock - Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### 3.2.showing daily percentage change in closing price using histogram.

```python
# Daily percentage change in closing price
stock_data['Daily_Return'] = stock_data['Close'].pct_change() * 100
# Histogram of daily returns
plt.figure(figsize=(10, 6))
plt.hist(stock_data['Daily_Return'].dropna(), bins=20, color='olive', alpha=0.7)
plt.title('Histogram of Daily Returns')
plt.xlabel('Daily Returns (%)')
plt.ylabel('Frequency')
plt.show()
```

### 3.3.showing rolling mean and standard deviation of closing price using line plot.

```python
# Rolling mean and standard deviation of closing price
rolling_window = 20
stock_data['Rolling_Mean'] = stock_data['Close'].rolling(window=rolling_window).mean()
stock_data['Rolling_Std'] = stock_data['Close'].rolling(window=rolling_window).std()
# Plotting the rolling mean and standard deviation
plt.figure(figsize=(12, 6))
plt.plot(stock_data['Date'], stock_data['Close'], label='Closing Price', color='blue')
plt.plot(stock_data['Date'], stock_data['Rolling_Mean'], label=f'{rolling_window}-Day Rolling Mean', color='orange')
plt.plot(stock_data['Date'], stock_data['Rolling_Std'], label=f'{rolling_window}-Day Rolling Std', color='red')
plt.title('Reliance Industries Stock - Rolling Mean and Standard Deviation')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### 3.4.showing the monthly average closing price using line plot.

```python
# Monthly average closing price
stock_data['Month'] = stock_data['Date'].dt.to_period('M')
monthly_avg = stock_data.groupby('Month')['Close'].mean()

# Plotting the monthly average closing price
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg.index.astype(str), monthly_avg.values, marker='o', linestyle='-', color='purple')
plt.title('Monthly Average Closing Price')
plt.xlabel('Month')
plt.ylabel('Average Closing Price')
plt.xticks(rotation=45)
plt.show()
```

### 3.5.showing closing prices by month using box plot.

### Bollinger Bands:-
#### Bollinger Bands are a technical analysis tool used in stock market data analysis to assess the volatility and potential price levels of a financial instrument, such as a stock.

### 3.10.showing bollinger bands using line plot.

```python
# Bollinger Bands
window = 20
stock_data['Rolling_Mean'] = stock_data['Close'].rolling(window=window).mean()
stock_data['Upper_Band'] = stock_data['Rolling_Mean'] + 2 * stock_data['Close'].rolling(window=window).std()
stock_data['Lower_Band'] = stock_data['Rolling_Mean'] - 2 * stock_data['Close'].rolling(window=window).std()

# Plotting Bollinger Bands
plt.figure(figsize=(12, 6))
plt.plot(stock_data['Date'], stock_data['Close'], label='Closing Price', color='blue')
plt.plot(stock_data['Date'], stock_data['Rolling_Mean'], label=f'{window}-Day Rolling Mean', color='orange')
plt.plot(stock_data['Date'], stock_data['Upper_Band'], label='Upper Bollinger Band', color='red', linestyle='--')
plt.plot(stock_data['Date'], stock_data['Lower_Band'], label='Lower Bollinger Band', color='green', linestyle='--')
plt.title('Reliance Industries Stock - Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### Moving Average Convergence Divergence (MACD):-
#### The Moving Average Convergence Divergence (MACD) is a popular technical indicator used in stock market data analysis to identify potential trends, reversals, and momentum in the price movements of a financial instrument, such as a stock. It is a trend-following momentum indicator that shows the relationship between two moving averages of a security's price.

### 3.12.showing moving average convergence divergence (MACD) using plotting MACD and signal line using line                 plot. 

```python
# Moving Average Convergence Divergence (MACD)
short_window = 12
long_window = 26

stock_data['Short_EMA'] = stock_data['Close'].ewm(span=short_window, adjust=False).mean()
stock_data['Long_EMA'] = stock_data['Close'].ewm(span=long_window, adjust=False).mean()

stock_data['MACD'] = stock_data['Short_EMA'] - stock_data['Long_EMA']
stock_data['Signal_Line'] = stock_data['MACD'].ewm(span=9, adjust=False).mean()

# Plotting MACD and Signal Line
plt.figure(figsize=(12, 6))
plt.plot(stock_data['Date'], stock_data['MACD'], label='MACD', color='blue')
plt.plot(stock_data['Date'], stock_data['Signal_Line'], label='Signal Line', color='orange')
plt.title('Reliance Industries Stock - MACD and Signal Line')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### Average True Range (ATR):-
#### The Average True Range is a valuable tool for risk management and position sizing in trading. It helps market participants adapt their strategies to different market conditions and make more informed decisions based on the current level of volatility.

### 3.13.showing average true range(ATR) using line plot.

```python
# Average True Range (ATR)
window = 14
stock_data['TR'] = np.maximum.reduce([stock_data['High'] - stock_data['Low'],
                                      abs(stock_data['High'] - stock_data['Close'].shift(1)),
                                      abs(stock_data['Low'] - stock_data['Close'].shift(1))])

stock_data['ATR'] = stock_data['TR'].rolling(window=window).mean()

# Plotting ATR
plt.figure(figsize=(12, 6))
plt.plot(stock_data['Date'], stock_data['ATR'], label=f'{window}-Day ATR', color='green')
plt.title('Reliance Industries Stock - Average True Range (ATR)')
plt.xlabel('Date')
plt.ylabel('ATR')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
